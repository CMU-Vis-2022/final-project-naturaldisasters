```python
import pandas as pd
nd_m = pd.read_csv('/Users/zhangweiyuan/Desktop/dvfinal/metalhealth.csv')
nd_m
```

```python
import matplotlib.pyplot as plt
import numpy as np

y=[]
x=[]
y1=0
for i in range(0,len(nd_m)):
    if nd_m.iloc[i][0] == 'Resident':
        y1 = nd_m.iloc[i][1] + y1
        y.append(y1)
        x.append(nd_m.iloc[i][2])
x,y
```

```python
z=[]
z1=0
for i in range(0,len(nd_m)):
    if nd_m.iloc[i][0] == 'Patients':
        z1 = nd_m.iloc[i][1] + z1
        z.append(z1)
z
```

```python
m=[]
m2=[]
m1=0
for i in range(0,len(nd_m)):
    if nd_m.iloc[i][0] == 'Employee':
        m1 = nd_m.iloc[i][1] + m1
        m.append(m1)
        m2.append(nd_m.iloc[i][2])
m
```

```python
n=[]
n1=0
for i in range(0,len(nd_m)):
    if nd_m.iloc[i][0] == 'Students':
        n1 = nd_m.iloc[i][1] + n1
        n.append(n1)
n
```

```python
plt.figure(figsize=(10,8), dpi=150)
plt.fill_between(x,y,alpha=0.4)
plt.fill_between(m2,z,alpha=0.4)
plt.fill_between(m2,n,alpha=0.4)
plt.fill_between(m2,m,alpha=0.4)
plt.legend(['Residents','Patients','Employee','Students'], loc='upper left')
plt.xlabel('Months')
plt.ylabel('Number of People(in thousands)')
plt.title('The number of people has mental issues after GEJE')
plt.text(0.5,-1,'Note: Observations were sampled independently and mutually exclusive', va='top',color='r')
plt.show()
```
