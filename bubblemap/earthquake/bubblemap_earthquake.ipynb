```python
nd_e = nd[nd['Disaster Type'] == 'Earthquake']
nd_e = nd_e.sort_values(by='Country')
nd_e
```

```python
nd_e1=nd_e.groupby(nd_e['Country'])['Total Affected'].sum()
nd_e1=nd_e1.tolist()
# nd_e1
```

```python
nd_e2 = nd_e['Country'].unique()
nd_e2.sort()
nd_e2=nd_e2.tolist()
# nd_e2
```

```python
nd_e3=nd_e['Latitude'].drop_duplicates().values.tolist()
# nd_e3
```

```python
nd_e4=nd_e['Longitude'].drop_duplicates().values.tolist()
# nd_e4
```

```python
nd_bubble = pd.DataFrame([nd_e2, nd_e3, nd_e4, nd_e1])
nd_bubble=nd_bubble.T
nd_bubble.columns = ['Country','Latitude','Longitude','Values']
nd_bubble
```

```python
import folium
m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=3)

for i in range(0,len(nd_bubble)):
   folium.Circle(
      location=[nd_bubble.iloc[i][1], nd_bubble.iloc[i][2]],
      popup=nd_bubble.iloc[i][0],
      radius=float(nd_bubble.iloc[i][3])/150,
      color='#69b3a2',
      fill=True,
      fill_color='#69b3a2'
   ).add_to(m)

# m
```
