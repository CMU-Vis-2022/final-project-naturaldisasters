```python
import pandas as pd
nd = pd.read_csv('/Users/zhangweiyuan/Desktop/dvfinal/disaster-bubblemap.csv')
nd
```

```python
nd_f = nd[nd['Disaster Type'] == 'Flood']
nd_f = nd_f.sort_values(by='Country')
nd_f
```

```python
nd_f1=nd_f.groupby(nd_f['Country'])['Total Affected'].sum()
nd_f1=nd_f1.tolist()
# nd_e1

nd_f2 = nd_f['Country'].unique()
nd_f2.sort()
nd_f2=nd_f2.tolist()
# nd_e2

nd_f3=nd_f['Latitude'].drop_duplicates().values.tolist()

nd_f4=nd_f['Longitude'].drop_duplicates().values.tolist()
```

```python
nd_bubbleflood = pd.DataFrame([nd_f2, nd_f1, nd_f3, nd_f4])
nd_bubbleflood=nd_bubbleflood.T
nd_bubbleflood.columns = ['Country','Values','Latitude','Longitude']
nd_bubbleflood
```

```python
from pyecharts.globals import GeoType
from pyecharts.globals import ChartType, SymbolType
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python
a1_listone=[]
datapairflood=[]
for i in range(0,len(nd_bubbleflood)):
    a1_listone=[nd_bubbleflood.iloc[i][0],nd_bubbleflood.iloc[i][1]]
    a=tuple(a1_listone)
    datapairflood.append(a)
#以上部分是为了产生元组列表，大家可以参考

worldmap1=Map(init_opts=opts.InitOpts(width="1200px",height="600px",bg_color='white'))
worldmap1.add(" ",data_pair=datapairflood,maptype="world",is_map_symbol_show=False)
worldmap1.set_series_opts(label_opts=opts.LabelOpts(is_show=True, font_size=5))
worldmap1.set_global_opts(
    title_opts=opts.TitleOpts(title="Total Affected People by Flood by Country"),
    legend_opts=opts.LegendOpts(is_show=False),
    visualmap_opts=opts.VisualMapOpts(min_=0,max_=20000000,orient="horizontal",type_ ="color",
                                                        #    range_color=['#FFFFFF','#E0FFFF','#AFEEEE','#ADD8E6','#87CEEB',
                                                        #    '#6495ED','#4682B4'],
                                                           pos_left=550,pos_bottom=60,item_width=15,item_height=200))
worldmap1.render()
```
