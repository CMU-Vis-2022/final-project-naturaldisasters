```python
import pandas as pd
nd = pd.read_csv('/Users/zhangweiyuan/Desktop/dvfinal/disaster-bubblemap.csv')
nd
```

```python
nd_s = nd[nd['Disaster Type'] == 'Storm']
nd_s = nd_s.sort_values(by='Country')
nd_s
```

```python
nd_s1=nd_s.groupby(nd_s['Country'])['Total Affected'].sum()
nd_s1=nd_s1.tolist()
# nd_e1

nd_s2 = nd_s['Country'].unique()
nd_s2.sort()
nd_s2=nd_s2.tolist()
# nd_e2

nd_s3=nd_s['Latitude'].drop_duplicates().values.tolist()

nd_s4=nd_s['Longitude'].drop_duplicates().values.tolist()
```

```python
nd_bubblestorm = pd.DataFrame([nd_s2, nd_s1, nd_s3, nd_s4])
nd_bubblestorm=nd_bubblestorm.T
nd_bubblestorm.columns = ['Country','Values','Latitude','Longitude']
nd_bubblestorm
```

```python
from pyecharts.globals import GeoType
from pyecharts.globals import ChartType, SymbolType
from pyecharts.charts import Geo
from pyecharts import options as opts
from pyecharts.charts import Map
```

```python
a2_listone=[]
datapairstorm=[]
for i in range(0,len(nd_bubblestorm)):
    a2_listone=[nd_bubblestorm.iloc[i][0],nd_bubblestorm.iloc[i][1]]
    a=tuple(a2_listone)
    datapairstorm.append(a)
#以上部分是为了产生元组列表，大家可以参考

worldmap2=Map(init_opts=opts.InitOpts(width="1200px",height="600px",bg_color='white'))
worldmap2.add(" ",data_pair=datapairstorm,maptype="world",is_map_symbol_show=False)
worldmap2.set_series_opts(label_opts=opts.LabelOpts(is_show=True, font_size=5))
worldmap2.set_global_opts(
    title_opts=opts.TitleOpts(title="Total Affected People by Storm by Country"),
    legend_opts=opts.LegendOpts(is_show=False),
    visualmap_opts=opts.VisualMapOpts(min_=0,max_=20000000,orient="horizontal",type_ ="color",
                                                        #    range_color=['#FFFFFF','#E0FFFF','#AFEEEE','#ADD8E6','#87CEEB',
                                                        #    '#6495ED','#4682B4'],
                                                           pos_left=550,pos_bottom=60,item_width=15,item_height=200))
worldmap2.render()
```
